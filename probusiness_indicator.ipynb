```python
import os
import pandas as pd

# Carga las listas de palabras desde los archivos CSV
palabras = pd.read_excel('/content/listapalabrasempresa.xlsx')



palabras


```

```python
lista1 = palabras.loc[0:49,'palabra50'].tolist()
lista2 = palabras['palabra173'].tolist()
lista1
```

```python
# Directorio donde están los archivos .txt
directorio = '/content/speeches'

# Crea un dataframe vacío
df = pd.DataFrame(columns=['archivo', 'ocurrencias_lista50', 'ocurrencias_lista173'])

# Itera sobre todos los archivos .txt en el directorio
for filename in os.listdir(directorio):
    if filename.endswith(".txt"):
        with open(os.path.join(directorio, filename), 'r', encoding='utf-8') as f:
            texto = f.read()

        # Cuenta las ocurrencias de todas las palabras en cada lista y las guarda en un diccionario
        data = {}
        data['ocurrencias_lista50'] = sum(texto.count(palabra) for palabra in lista1)
        data['ocurrencias_lista173'] = sum(texto.count(palabra) for palabra in lista2)
        data['palabras']=len(texto.split())

        # Añade el nombre del archivo sin la extensión .txt al diccionario
        data['archivo'] = os.path.splitext(filename)[0]

        # Añade los datos del archivo al dataframe
        df = df.append(data, ignore_index=True)

# Guarda el dataframe como un archivo CSV
df.to_csv('/content/resultados.csv', index=False)
```
